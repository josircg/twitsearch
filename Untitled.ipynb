```python
import json

# http://docs.tweepy.org/en/v3.5.0/cursor_tutorial.html
# http://docs.tweepy.org/en/v3.5.0/streaming_how_to.html
# https://pythonprogramming.net/twitter-api-streaming-tweets-python-tutorial/

class SimpleListener(tweepy.StreamListener):
    def on_status(self, status): 
        #code to run each time the stream receives a status
        print(status.text)
  
    def on_data(self, status):
        #code to run each time you receive some data (direct message, delete, profile update, status,...)
        all_data = json.loads(status)
        filename = 'data/%s.json' % all_data['id']
        arquivo = open(filename,'w')
        arquivo.write(status)
        arquivo.close()
        print('%s saved' % filename)

    def on_error(self, status_code):
        #code to run each time an error is received
        if status_code == 420:
            return False
        else:
            return True
```

```python
#initialize the stream

listener = SimpleListener()
tweepy_stream = tweepy.Stream(auth = api.auth, listener=listener)
tweepy_stream.filter(track=['1964'])
```

```python
import os, glob
from collections import Counter

tot_retwit = 0
tot_users = Counter()
files = [f for f in glob.glob("*.json", recursive=False)]
for filename in files:
    arquivo = open(filename,'r')
    status = arquivo.read()
    if status:
        tweet = json.loads(status)
        tot_users[ tweet['user']['screen_name'] ] += 1
        if 'retweeted_status' in tweet:
            tot_users[ tweet['retweeted_status']['user']['screen_name'] ] += 1
            tot_retwit += 1
    arquivo.close()

print(tot_retwit)
```
